```python
import numpy as np
import matplotlib.pyplot as plt
import math

print("Para hacer una transformacion lineal a X=e^((y-b)/a) basta con despejar ¨y¨ al aplicar ln a ambos lados, quedando y=a.ln(x)+ b")

# Datos de muestra
x = np.array( [math.log(1),math.log(2),math.log( 3), math.log(4), math.log(5)])
y = np.array([0.5, 2, 2.9, 3.5, 4])

print(f"Entonces para unos valores de ln(x)={x} y valores de y={y}")

# Cálculo de la pendiente (m) y la ordenada al origen (b)
n = len(x)
m = (n * np.sum(x * y) - np.sum(x) * np.sum(y)) / (n * np.sum(x**2) - (np.sum(x))**2)
b = (np.sum(y) - m * np.sum(x)) / n

# Ecuación de regresión lineal
regression_line = m * x + b

# Visualización de los resultados
plt.scatter(x, y, label='Datos de muestra')
plt.plot(x, regression_line, color='red', label='Regresión lineal')
plt.xlabel('Variable Independiente ln(x)')
plt.ylabel('Variable Dependiente (y)')
plt.title('Regresión por Mínimos Cuadrados')
plt.legend()
plt.show()

# Imprimir la ecuación de regresión
print(f"La ecuación de regresión es: y = {m}x + {b}")
print(f"Por lo que el valor de a es {m} y el de b es {b}")
print(f"Y para un x = {2.6} se tendra un y = {math.exp((2.6-b)/m)}")
```
