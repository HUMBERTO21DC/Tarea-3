```python
import numpy as np
import matplotlib.pyplot as plt
from math import exp

#Para determinar la constante de resistencia c, se busca la ecuacion fisica correspondiente que es "v=(m*g)/c) * (1 - np.exp (-1*c*t/m)". Se despeja v, se insertan los valores datos, se grafica y el punto que tenga y=0 el valor de c sera el correspondiente valor de x en ese punto.

#datos
m= int (68.1)
g= int (9.81)
t =10
v = 40



def f(c):
    return  ((( (m*g)/c) * (1 - np.exp (-1*c*t/m)))-v)


c_valores = np.linspace(10,15,20)
f_values = f(c_valores)

plt.plot(c_valores, f_values, color='blue',  label='f(x) vs c')
plt.xlabel('Eje c')
plt.ylabel('Eje f(c)')
plt.title ('F(c) vs c')
plt.annotate ('Valor de la constante de resistencia c =13.058023275597940', xy=(13.058023275597940,0))
plt.axhline (0, color='black')
plt.axvline (13.058023275597940, color='black')
plt.show()
```

```python
import numpy as np

matriz = np.array ([[0.3,0.52,1],
                    [0.5,1,1.9],
                    [0.1,0.3,0.5]])
det_m = np.linalg.det(matriz)
print (f"Tenemos la matriz: {matriz} con su determinante {det_m}")


Resulados = np.array ([-0.01,0.67,-0.44])
print(f"Con resultados {Resulados}")

print("Si se cambia los valores de la primera columna de la matriz con los valores de los resultados el determinante es:")
matriz_x = np.array ([[-0.01,0.52,1],
                      [0.67,1,1.9],
                      [-0.44,0.3,0.5]])
det_x = np.linalg.det(matriz_x)
print(det_x)

print("Si se cambia los valores de la segunda columna de la matriz con los valores de los resultados el determinante es:")
matriz_y = np.array ([[0.3,-0.01,1],
                    [0.5,0.67,1.9],
                    [0.1,-0.44,0.5]])
det_y = np.linalg.det(matriz_y)
print(det_y)

print("Si se cambia los valores de la tercera columna de la matriz con los valores de los resultados el determinante es:")
matriz_z = np.array ([[0.3,0.52,-0.01],
                    [0.5,1,0.67],
                    [0.1,0.3,-0.44]])
det_z = np.linalg.det(matriz_z)
print(det_z)

print ("Entonces...")

valor_X = det_x / det_m
print(f"x es igual a {valor_X}")

valor_y = det_y / det_m
print(f"y es igual a {valor_y}")

valor_z = det_z / det_m
print(f"z es igual a {valor_z}")
```
